### LR UCI Dataset

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sb
import statistics as state
```

```python
#df=pd.read_csv("glass.csv")
ColumnNames=['Id','RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe', 'Type']
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/glass/glass.data', names=ColumnNames)
df.drop(["Id"], axis=1, inplace=True)
df.to_csv('glass_lr.csv')
df.head(214)
```

### setting the matrixes

```python
X =df.iloc[:,0:9]
print(X)
Y = df.iloc[:,9:10]
print(Y)
```

### Calculating Slope,Intercept & R-square

```python
regr = linear_model.LinearRegression()
regr.fit(X, Y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
rsquare = regr.score(X,Y)*100
print('R-Square: %.2f' %rsquare + '%')
```

### Predicted Value of Class-Type 

```python
Type=regr.predict(X)
print(Type.T)
print(np.shape(Type))
```
